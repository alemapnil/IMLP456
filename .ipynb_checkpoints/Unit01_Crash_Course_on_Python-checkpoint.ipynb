### 01 - 奇偶數辨別（Odd or even）
#### 當使用者輸入一個的數字，它能夠辨別奇偶，並輸出檢驗結果給使用者。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一個數字 <br>
輸出(Output):奇數或偶數

* 進階: 可以限制輸入一個介於一定範圍（例如 1 到 1000）的數字

````python
def odd_even():
    try:
        input_num = int(input("Please enter an integer: "))
        
        if input_num < 1 or input_num > 1000:
            raise TypeError()
    except ValueError:
        print('Please enter an integer')
    except TypeError:
        print('Integer should be ranged from 1 to 1000')
    else:
        if input_num%2 == 1:
            print("It's an odd")
        else:
            print("It's an even")

odd_even()
````

### 02 - 填字遊戲（Mad libs game）
#### 使用者輸入任何字，它可以是名詞、形容詞、動詞、代名詞等。程式得到文字後，可自行排列，組成成一個模板故事。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一連串依據說明輸入對應文字 <br>
輸出(Output): 模板故事

* 進階: 可以讓程式重複執行

````python
def mad_libs_game():
    name = input("What's your name ? ")
    nation = input("Where are you from ? ")
    job  = input("What job title you want to be ? ")
    industry = input("What kind of industry you want to devote to ? ")
    print('----------')

    lines = f"""
    Hello ~ {name}, who is from {nation}. Good to see you.\n
    I've heard that you strive for a post of {job}, and you've made lots of efforts into {industry}.\n
    It's nice to see you have an objective and go for it no matter how hard you've been through.\n
    Just keep faith in yourself. Little by little, step by step.\n
    You'll find how much power inside of you.\n"""
    print(lines)

    while True:
        ask  = input("Do you want to continue MAD LIBS game? (yes/no)")
        if ask == 'yes':
            mad_libs_game()
            return
        elif ask == 'no':
            print('Bye Bye !')
            break

mad_libs_game()
````

````python
def guess():
    import random
    ans = random.randint(1,100)
    count = 0
    high,low = 100,1
    while True:
        try:
            input_num = int(input(f"Guess an integer from {low} to {high}: "))
            print("")
            if input_num < 1 or input_num > 100:
                raise TypeError()
        except ValueError:
            print('Please enter an integer')
        except TypeError:
            print('Integer should be ranged from 1 to 100')
        else:
            count += 1
            if input_num == ans:
                print(f"Bingo ! The answear is {ans}.")
                break
            elif input_num > ans:
                print(f"Too large. You've guessed {count} times.")
                if high > input_num or high == input_num:
                    high = input_num
                ask = input(f"Do you want to quit or not (q/no)? ")
                if ask == 'q':
                    print('Bye Bye !')
                    break
            elif input_num < ans:
                print(f"Too small. You've guessed {count} times.")
                if low < input_num or low == input_num:
                    low = input_num
                ask = input(f"Do you want to quit or not (q/no)? ")
                if ask == 'q':
                    print('Bye Bye !')
                    break

guess()
````

### 06 - 及格名單（Pass list）
#### 請使用集合功能來完成各科級名單的判斷
```
米花市帝丹小學一年级B班正舉辦期中考試
數學及格的有：柯南、灰原、步美、美環、光彦
英文及格的有：柯南、灰原、丸尾、野口、步美
以上已列出全班所有人
```

>請分別列出<br>
數學及格但英文不及格的同學名單<br>
數學不及格但英文及格的同學名單<br>
兩者皆及格名單<br>
* Hint:差集(減法)、交集


````python
def passlist():
    math_pass = {'柯南','灰原','步美','美環','光彦'}
    eng_pass = {'柯南','灰原','丸尾','野口','步美'}

    all = [i for i in math_pass] #全班同學

    for i in eng_pass:
        if i not in all:
            all.append(i)

    math_pass_not_eng, math_not_eng_pass, both_pass = [], [], []   
    for i in all:
        if i in math_pass and i not in eng_pass:
            math_pass_not_eng.append(i)
        elif i not in math_pass and i in eng_pass:
            math_not_eng_pass.append(i)
    
    for i in all:
        if i in math_pass and i in eng_pass:
            both_pass.append(i)

    print(f"Pass Math but fail in English: {math_pass_not_eng}")
    print(f"Fail in Math but pass English: {math_not_eng_pass}")
    print(f"Both pass: {both_pass}")

passlist()
````

* 進階: 把file的學號抓出來，另存成一個新的csv檔

````python
def filename():
    import os
    current_path = os.getcwd()
    path = f"{current_path}/testfile"
    filelist = os.listdir(path)
    file_number = []
    for i in filelist:
        if "." not in i:
            print(f"I'm a directory: {i}")
        else:
            print(f"I'm a file: {i}")
            file_number.append(i.split(".")[0])
            
    print(f"I'm the list of number of files: {file_number}")

filename()
````

### 08 - 資料夾管理(shutil)
#### 說明 ####

```
在目前所在的目錄下建立一files資料夾

令使用者輸入一數字N，並在files資料夾中建立f1, f2… fN等N個資料夾後列出files的資料夾內容

將files資料夾裡的f1資料夾重新命名成folder1後再列出files的資料夾內容

移除files資料夾中的folder1後再列出files的資料夾內容

最後移除files資料夾
※須先退出files資料夾(os.chdir(../)) 
```


````python
def folder():
    import os, shutil
    os.makedirs("files", exist_ok=True) 
    
    print("The folder of files has created.")
    while True:
        try:
            N = int(input("Please enter an integer: "))
            break
        except ValueError:
            print("It\'s not an integer")
    for i in range(1, N + 1):
        os.makedirs(f"files/f{i}", exist_ok=True) 
    
    current_path = os.getcwd()
    srearch_path = f"{current_path}/files"

    def key(n):
        return int(n[1:])
    
    dirs_list = sorted(os.listdir(srearch_path),key=key)
    print(dirs_list)
    os.rename('files/f1','files/folder1')
    print(os.listdir(srearch_path))
    shutil.rmtree('files/folder1')
    dirs_list = sorted(os.listdir(srearch_path),key=key)
    print(dirs_list)
    shutil.rmtree('files')

folder()
````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python
def palindrome():
    raw = input("Please enter a word, phrase: ")
    count = 0 
    last_index = len(raw)-1

    if len(raw)%2 == 1: # odd
        mid_index = len(raw)//2

        for I,i in enumerate(raw):
            if I < mid_index and i != raw[last_index-I]: # not palindrome
                count += 1

    else: #even
        for I,i in enumerate(raw):
            if i != raw[last_index-I]:# not palindrome
                count += 1 

    if count == 0:
        print("The text you have entered is a palindrome ! ")
    else:
        print("The text you have entered is not a palindrome.")

palindrome()
````

### 10 - 總成績計算

````python
def score():
    import pandas as pd
    import os

    current = os.getcwd()
    # english_list經驗測 姓名為唯一值
    df1 = pd.read_csv(rf"{current}\Unit01\Unit01\data\english_list.csv")

    # math_list經驗測 姓名為唯一值
    df2 = pd.read_csv(rf"{current}\Unit01\Unit01\data\math_list.csv")

    result = pd.merge(df1, df2, on="姓名")

    result["score"] = result['英文']+result['數學']

    result = result.drop(columns=['英文', '數學'])

    result.to_csv('Score.csv', index=False,header=False)  

score()
````
