## 1. 陣列導向 101

科學計算一個很核心的概念叫 "array oriented" 的寫法。Array 是 `numpy` 標準的資料結構, 和 list 很像, 但就差了那麼一點點。而這一點點讓我們在計算上是無比的方便。

### 【暖身】 計算平均

某位同學期中考各科成績如下, 請幫他計算成績。

    grades = [77, 85, 56, 90, 66]
    
請計算平均。

```python
grades = [77, 85, 56, 90, 66]
s = sum(grades)
length = len(grades)
avg = s/length
avg = "{:.2f}".format(avg)
print(avg)  
```

### 【示範】陣列導向

```python
arr = np.array(grades)
mean = np.mean(arr)
print(mean)
```

最大值

```python
maxium = np.max(arr)
print(maxium)
```

標準差

```python
sd = np.std(arr)
print(sd)
```

### 【暖身】 換算匯率

假設今天我想查查號稱 Pentax 三公主的 31mm, 43mm, 77mm 三隻 limited 鏡頭在美國賣多少。於是我去 B&H 查了他們的價格分別是:

    prices = [1096.95, 596.95, 896.95]
    
我又查了 Google 匯率 1 美金為 31.71 元。請把三支鏡頭的價格換算為台幣。

```python
prices = [1096.95, 596.95, 896.95]
currency = 31.71
twd_p = []

for i in prices:
    convert = "{:.4f}".format(i*currency)
    twd_p.append(float(convert))

print(twd_p)
```

先不管這實在有夠醜的數字, 我們要記得在科學計算中:

### 儘可能不要使用迴圈

這可能嗎?

### 【示範】陣列換算匯率

```python
arr = np.array(prices)
twd_p_arr = np.round(arr*currency, decimals=4)
print(twd_p_arr)
```

哦哦, 傑克, 這太神奇了!

## 2. 其實 array 還有很多功能

### 【練習】 成績計算

一位老師成績這樣算的:

* 平時成績 20%
* 期中考   35%
* 期未考   45%

有位同學

* 平時成績 85 分
* 期中 70 分
* 期末 80 分

這位同學的學期成績是多少?

```python
grades = np.array([85,70,80])
weights = np.array([0.2, 0.35, 0.45])
```

這還不是我們要的最終成績啊!

```python
final_grades = np.average(grades, weights = weights)
print(final_grades)
```

### 【提示】 array 還有很多函數可以用

可以先打入

    weighted_grades.
    
先不要按 `enter` 或 `shift-enter`, 而是按 `tab`...


```python
final_grades = np.average(grades, weights = weights)
print(final_grades)
```

### 【技巧】 一行完成成績計算

```python
ans = np.dot(grades,weights)
print(ans)
```

## 8. `NumPy` 的 `zip` 和 `unzip`

之前我們介紹 list 可以用 `zip` 和 `unzip` (其實還是 `zip`) 做到的資料格式變換, 在 array 中怎麼做呢?

![zip and unzip](images/zip.png)

### 【重點】array 的 `zip`

![array zip](images/arrzip.png)

```python
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])
```

```python
# np.c_[x] 將一維的 x 變成二維的欄向量
# np.c_[y] 將一維的 y 變成二維的欄向量
X = np.c_[x,y] # 水平堆疊兩個欄向量
X
```

```python
# np.r_[x] 不會自動增加維度，因為 x 已經是一維， r_ 預設就是沿著 axis=0 垂直連接元素
np.r_[x,y]
```

### 【重點】array 的 `unzip`

這裡其實只需要用到 array 的切割法...

![array zip](images/arrunzip.png)

```python
X[:,0]
```

```python
X[:,1]
```
